## Описание данных

Файл `visits_log.csv` хранит лог сервера с информацией о посещениях сайта, `orders_log.csv` — информацию о заказах, а `costs.csv` — информацию о расходах на маркетинг.

**Структура `visits_log.csv`**

- `Uid` — уникальный идентификатор пользователя,
- `Device` — категория устройства пользователя,
- `Start Ts` — дата и время начала сессии,
- `End Ts` — дата и время окончания сессии,
- `Source Id` — идентификатор источника перехода на сайт.

**Структура `orders_log.csv`**

- `Uid` — уникальный идентификатор пользователя,
- `Buy Ts` — дата и время заказа,
- `Revenue` — сумма заказа.

**Структура `costs.csv`**

- `source_id` — идентификатор рекламного источника,
- `dt` — дата проведения рекламной кампании,
- `costs` — расходы на эту кампанию.

#### Количество уникальных пользователей в день/неделю/месяц

```python
visits['start_year']  = visits['start_ts'].dt.year
visits['start_month'] = visits['start_ts'].dt.month
visits['start_week']  = visits['start_ts'].dt.week
visits['start_date'] = visits['start_ts'].dt.date
visits.head()
```

```python
dau_total = visits.groupby('start_date').agg({'uid': 'nunique'}).mean()
wau_total = (
    visits.groupby(['start_year', 'start_week'])
    .agg({'uid': 'nunique'})
    .mean()
)
mau_total = (
    visits.groupby(['start_year', 'start_month'])
    .agg({'uid': 'nunique'})
    .mean()
)

print('Среднее количество уникальных пользователей в день:', int(dau_total))
print('Среднее количество уникальных пользователей в неделю:', int(wau_total))
print('Среднее количество уникальных пользователей в месяц:', int(mau_total))
```

#### Когортный анализ и Retention Rate

Добавим столбец с датой первого визита:

```python
first_visit_date_by_customers = visits.groupby('uid')['start_ts'].min()
first_visit_date_by_customers.name = 'first_visit_date'
visits = visits.join(first_visit_date_by_customers, on='uid')
visits
```

Выделим месяцы:

```python
visits['first_visit_month'] = visits['first_visit_date'].astype('datetime64[M]')
visits['visit_month'] = visits['start_ts'].astype('datetime64[M]')
```

```python
visits
```

Подсчитаем lifetime:

```python
visits['lifetime'] = ((visits['visit_month'] - visits['first_visit_month']) / np.timedelta64(1,'M')).round().astype('int')
```

Сформируем когорты:

```python
visit_cohorts = visits.groupby(['first_visit_month','lifetime']).agg({'uid':'nunique'}).reset_index()
visit_cohorts['first_visit_month'] = visit_cohorts['first_visit_month'].dt.strftime('%Y-%m') 
```

Создадим датафрейм, в котором для каждой когорты будет отображено число пользователей на нулевой месяц lifetime:

```python
initial_users_count = visit_cohorts[visit_cohorts['lifetime'] == 0][['first_visit_month', 'uid']]
initial_users_count = initial_users_count.rename(columns={'uid':'initial_users'})
```

Объединим датафреймы:

```python
visit_cohorts = visit_cohorts.merge(initial_users_count, on='first_visit_month')
```

Рассчитаем показатель Retention Rate в отдельном столбце и выведем сводную таблицу, которая отобразит изменение *retention* по когортам в зависимости от *lifetime*:

```python
visit_cohorts['retention'] = (visit_cohorts['uid']/visit_cohorts['initial_users'])
retention_pivot = visit_cohorts.pivot_table(index='first_visit_month', columns='lifetime', 
                          values='retention', aggfunc='sum')
```

```python
retention_pivot
```

Построим тепловую карту:

```python
plt.figure(figsize=(12, 8))
plt.title('Cohorts: Retention Rate')
sns.heatmap(retention_pivot, annot=True, fmt='.1%', linewidths=0.05, linecolor='white', cmap='mako', vmax=0.2)
plt.show()
```

```python
rr_mean = retention_pivot[1].mean() * 100
print('Средний Retention Rate на второй месяц жизни когорт: {:.1f}%'.format(rr_mean))
```

**Вывод:** Совсем небольшая часть пользователей возвращается после первого месяца, далее retention rate постепенно уменьшается, но незначительно, а в некоторых когортах (июнь и июль 2017) даже немного увеличивается. Самый высокий retention rate у сентябрьской когорты (возможно это связано, например, с открытием нового театрального сезона).
В целом получается, что большая часть людей пользуется сервисом единоразово.

### Метрики электронной коммерции

#### Среднее время с момента первого посещения до совершения покупки

#### Среднее количество покупок на одного покупателя за период

Добавим столбец с датой первой покупки:

```python
first_purchase_date_by_customers = orders.groupby('uid')['buy_ts'].min()
first_purchase_date_by_customers.name = 'first_purchase_date'
purchases = orders.join(first_purchase_date_by_customers, on='uid')
purchases
```

Выделим месяцы:

```python
purchases['first_purchase_month'] = purchases['first_purchase_date'].astype('datetime64[M]')
purchases['purchase_month'] = purchases['buy_ts'].astype('datetime64[M]')
purchases
```

Посчитаем lifetime:

```python
purchases['lifetime'] = ((purchases['purchase_month'] - purchases['first_purchase_month']) / np.timedelta64(1,'M')).round().astype('int')
```

Сформируем когорты:

```python
purchases_cohorts = purchases.groupby(['first_purchase_month', 'lifetime']).agg({'uid':['count', 'nunique']}).reset_index()
purchases_cohorts.columns = ['first_purchase_month', 'lifetime', 'n_orders', 'n_users']
```

```python
purchases_cohorts['first_purchase_month'] = purchases_cohorts['first_purchase_month'].dt.strftime('%Y-%m') 
```

Рассчитаем среднее количество заказов на человека и сформируем сводную таблицу:

```python
purchases_cohorts['orders_per_user'] = purchases_cohorts['n_orders'] / purchases_cohorts['n_users']

purchases_cohorts_pivot = purchases_cohorts.pivot_table(index = 'first_purchase_month', columns = 'lifetime', values = 'orders_per_user', aggfunc = 'mean')
purchases_cohorts_pivot['mean'] = purchases_cohorts_pivot.mean(axis=1)
```

```python
purchases_cohorts_pivot
```

Построим тепловую карту:

```python
plt.figure(figsize=(12, 8))
plt.title('Cohorts: average number of purchases per user')
sns.heatmap(purchases_cohorts_pivot, annot=True, fmt='.1f', linewidths=0.05, linecolor='white')
plt.show()
```
